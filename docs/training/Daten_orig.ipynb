````python
from IPython.display import display, Markdown
display(Markdown("# Daten-Schulung"))

if "_weiter_repeat" not in globals():
    _weiter_repeat = None
if "_weiter_seen" not in globals():
    _weiter_seen = set()

def weiter(lösung):
    "Hilfsfunktion, die uns jeweils die nächste Übung laden wird"
    global _weiter_seen, _weiter_repeat
    lösung = "\n".join(str(lösung).splitlines()[:3])
    print(lösung + "\n")
    from hashlib import sha256
    import re
    h = sha256(lösung.replace(" ", "").lower().encode()).hexdigest()
    if "_steps" not in globals():
        return print(h)
    if  h in _steps:
        exercise = _steps[h]
    else:
        return print("Lösung falsch")
    instruction, code, hidden = exercise
    display(Markdown(instruction))
    if hidden:
        exec(hidden, globals())
    if lösung in _weiter_seen and lösung != _weiter_repeat:
        _weiter_repeat = lösung
        print("Code-Zelle bereits erzeugt, nochmal ausführen um trotzdem zu erzeugen")
    else:
        _weiter_seen.add(lösung)
        _weiter_repeat = None
        from IPython import get_ipython
        get_ipython().payload_manager.write_payload(dict(
            source='set_next_input',
            text=re.sub("^(.*) #\\? (.*)$", f"\\{1 if "_insert_solution" in globals() else 2}", code, flags=re.MULTILINE),
            replace=False,
        ))

def up(s, key="_"):
    return auth.s3.put_object(Bucket="guetersloh.dschulung", Key=f"snippet/{key}", Body=s.strip())

def down(key="_"):
    from IPython import get_ipython
    get_ipython().payload_manager.write_payload(dict(
        source='set_next_input',
        text=auth.s3.get_object(Bucket="guetersloh.dschulung", Key=f"snippet/{key}")['Body'].read().decode(),
        replace=True,
    ))

weiter("JupyterHub Grundlagen")
````

## Sensordaten

Sensordaten werden als Datenströme in die Plattform geladen.

[Dokumentation](https://docs.data-hub.local/schnittstellen/ingestor/)

Neben der Nutzbarkeit in [Diagrammen](https://superset.data-hub.local/explore/?datasource_type=table&datasource_id=1) und [SQL Lab](https://superset.data-hub.local/sqllab/?dbname=clickhouse&sql=SELECT%0A%20%20wkt((assumeNotNull(last_value(latitude))%2C%20assumeNotNull(last_value(longitude))))%20AS%20geometry%2C%0A%20%20last_value(time)%20AS%20t%2C%0A%20%20sensor_id%20AS%20sensor%2C%0A%20%20last_value(temperature_outside)%20AS%20aussentemperatur%0AFROM%20sensor_messages%0AGROUP%20BY%20sensor_id) können [Veröffentlichte Abfragen](https://docs.data-hub.local/berechtigungskonzept/#veroffentlichte-abfragen) erstellt werden,  
um die Daten als GeoJSON für z.B. Masterportal nutzbar zu machen.

````python
try:
    # falls noch da vom letzen Mal erst löschen
    auth.graphql.execute(gql("""
    mutation ($prefix: String!) {
      tenant(tenant: "guetersloh") {
        vizGroup(vizGroup: "dschulung") {
          deletePublishedQuery(publishedQuery: $prefix)
        }
      }
    }"""), variable_values=dict(prefix=prefix))
except:
    pass

r = auth.graphql.execute(gql("""
mutation ($prefix: String!, $sql: String!) {
  tenant(tenant: "guetersloh") {
    vizGroup(vizGroup: "dschulung") {
      createPublishedQuery(publishedQuery: $prefix, config: {sql: $sql}) {
        __typename
      }
    }
  }
}"""), variable_values=dict(prefix=prefix, sql="""
SELECT
  wkt((assumeNotNull(last_value(longitude)), assumeNotNull(last_value(latitude)))) AS geometry,
  last_value(time) AS t,
  sensor_id AS sensor,
  last_value(temperature_outside) AS aussentemperatur
FROM sensor_messages
GROUP BY sensor_id
"""))

weiter(r)
````
