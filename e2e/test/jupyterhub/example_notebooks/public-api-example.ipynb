```python
# Beispiel Jupyter-Notebook: Abfrage von Smart-City-Daten aus einer offenen API

import requests

# Funktion, um Daten von einer offenen API abzurufen
def fetch_smart_city_data(api_url):
    try:
        response = requests.get(api_url)
        response.raise_for_status()  # Überprüfen, ob der Abruf erfolgreich war
        return response.json()
    except requests.exceptions.RequestException as e:
        print(f"Fehler beim Abrufen der Daten: {e}")
        return None

def get_station_by_id(metadata ,station_id):
    for station in metadata["stations"]:
        if station["id"] == station_id:
            return station
    return None  # Gibt `None` zurück, wenn die ID nicht gefunden wird

# Beispiel-API für Smart-City-Daten (ohne Authentifizierung)
api_url = "https://api.data.gov.sg/v1/environment/air-temperature"

# Abrufen der Daten
data = fetch_smart_city_data(api_url)

# Überprüfen, ob Daten erfolgreich abgerufen wurden
if data:
    # Verarbeitung der Daten: Extrahieren von relevanten Informationen
    print("Es konnten Daten geladen werden");
    try:
        # Extrahieren der Einträge aus den API-Daten
        items = data["items"]
        metadata = data["metadata"]

        # Umwandeln in eine strukturierte Liste für bessere Übersichtlichkeit
        records = []

        for item in items:
            timestamp = item.get("timestamp")
            for reading in item.get("readings", []):
                station_id = reading.get("station_id")
                temperature = reading.get("value")
                station = get_station_by_id(metadata, reading["station_id"])
                records.append({
                    "timestamp": timestamp,
                    "station_name": station["name"],
                    "temperature": temperature
                })

        # Anzeigen der Daten in einer einfachen Textübersicht
        print("Datenvorschau:")
        for record in records[:5]:  # Begrenzen der Ausgabe auf die ersten 5 Einträge
            print(record)
    except Exception as e:
        print(f"Fehler bei der Verarbeitung der Daten: {e}")
else:
    print("Keine Daten verfügbar.")
```
